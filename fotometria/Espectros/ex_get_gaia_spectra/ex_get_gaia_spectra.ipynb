```python
from gaiaxpy import calibrate
```

```python
calibrated_spectra, sampling = calibrate([2341866690928009216])

calibrated_spectra
```

```python
from gaiaxpy import plot_spectra
from matplotlib import pyplot as plt
plt.rcParams["text.usetex"] = False

plot_spectra(calibrated_spectra, sampling=sampling, legend=False)
```

```python
import pandas as pd

x = sampling
y = calibrated_spectra.flux[0]

df = pd.DataFrame({'wavelength_[nm]': x, 'flux_[W/m²/nm]': y})
df.to_csv('../Gaia/gaia_spectra_2341866690928009216.csv', index=False)
```
